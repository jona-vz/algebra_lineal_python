```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
x = np.arange(-6,6)

y1 = 3*x+5
y2 = -x+3
y3 = 2*x+1
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```
